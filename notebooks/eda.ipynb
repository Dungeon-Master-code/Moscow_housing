–≠—Ç–æ—Ç –Ω–æ—É—Ç–±—É–∫ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ä–∞–∑–≤–µ–¥–æ—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö (EDA) –¥–ª—è –¥–∞—Ç–∞—Å–µ—Ç–∞ Moscow Housing Price Dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set(style='whitegrid', palette='muted', font_scale=1.1)
```

```python
df = pd.read_csv('/home/cat/projects/Moscow_housing/data/raw/data.csv')
print('–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞:', df.shape)
df.head()
```

```python
df.info()

df.describe().T
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Price'], bins=50, kde=True)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–Ω (Price)')
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(np.log1p(df['Price']), bins=50, kde=True)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–Ω (log(Price))')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(12,8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Number of rooms', y='Price', data=df)
plt.title('–¶–µ–Ω–∞ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —á–∏—Å–ª–∞ –∫–æ–º–Ω–∞—Ç')
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(x='Renovation', y='Price', data=df)
plt.title('–¶–µ–Ω–∞ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ç–∏–ø–∞ —Ä–µ–º–æ–Ω—Ç–∞')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Area', y='Price', data=df, alpha=0.5)
plt.title('–¶–µ–Ω–∞ vs –ü–ª–æ—â–∞–¥—å –∫–≤–∞—Ä—Ç–∏—Ä—ã')
plt.show()

plt.figure(figsize=(10,5))
sns.scatterplot(x='Minutes to metro', y='Price', data=df, alpha=0.5)
plt.title('–¶–µ–Ω–∞ vs –í—Ä–µ–º—è –¥–æ –º–µ—Ç—Ä–æ')
plt.show()
```
